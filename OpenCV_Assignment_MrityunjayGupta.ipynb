```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

**HISTOGRAM FOR OPENCV**

```python
img1 = cv2.imread("cat2.jpg")

histogram = cv2.calcHist([img1],[0], None,[256],[0,256])
plt.plot(histogram)
plt.show
```

**ADDING TEXT TO IMAGE**

```python
img2 = cv2.imread("cat2.jpg")

text = "What You Doin'?"
font = cv2.FONT_HERSHEY_PLAIN
org = (150,150)
fontscale = 15
color = (0,255,0)
thickness = 10

image_new = cv2.putText(img2,text,org,font,fontscale,color,thickness)

cv2_imshow(image_new)
```

**CIRCLE ON A IMAGE**

```python
img3 = cv2.imread("cat2.jpg")

img_circ = cv2.circle(img3, (200,200), 50, (0,255,0), 10)
cv2_imshow(img_circ)
```
